```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
twitter = pd.read_csv('~/LighthouseLabs-Final/Dataset/3. Twitter_Data/AAPL_2020-09-23_2020-10-07.csv', index_col=0, parse_dates=['timestamp'])
```

```python
twitter.set_index('timestamp',inplace=True)
```

```python
twitter['tweet_text']
```

```python
twitter2 = twitter.resample('30min').agg({'tweet_text':'size'})
```

```python
fig = plt.figure(figsize=(20,20))

price_ax = plt.subplot(6,1,1)
plt.plot(twitter2[:42])
price_ax = plt.subplot(6,1,2)
plt.plot(twitter2[48:90])
price_ax = plt.subplot(6,1,3)
plt.plot(twitter2[96:138])
price_ax = plt.subplot(6,1,4)
plt.plot(twitter2[240:282])
price_ax = plt.subplot(6,1,5)
plt.plot(twitter2[288:330])
price_ax = plt.subplot(6,1,6)
plt.plot(twitter2[336:378])

fig.subplots_adjust(hspace=0.2)

plt.savefig('daily_tweets.png')
```
